```python
import requests as req
import json
import pandas as pd
import numpy as np

kan = pd.read_csv('WK2000.csv',index_col=0)
```

```python
base = 'https://api.wanikani.com/v2'
token = ''
head = {'Authorization': f'Bearer {token}'}

#1-20, 21-40, 41-60 is best increments for reading all subjects (1000 limit)
#1-2, 13-24, 25-36, 37-48, 49-60 is best increments for reading all assignments (500 limit)

min_level = 13
max_level = 24

lvlstr=''
for l in range(min_level,max_level+1):
    lvlstr+=str(l)
    lvlstr+=','
lvlstr = lvlstr[:len(lvlstr)-1]

# endpoint = f'subjects?levels={lvlstr}&types=kanji'
endpoint = f'assignments?srs_stages=9&subject_type=kanji'

response = req.get(f'{base}/{endpoint}',headers=head)
res = json.loads(response.text)
print(len(res['data']))
```

```python
ids = [d['id'] for d in res['data']]
WKlvl = [d['data']['level'] for d in res['data']]
char = [d['data']['characters'] for d in res['data']]
imi = []
onyo=[]
kuyo=[]
for d in res['data']:
    temp_dat = d['data']
    on=[]
    kun=[]
    for r in temp_dat['readings']:
        if r['type'] == "onyomi":
            on.append(r['reading'])
        if r['type'] == "kunyomi":
            kun.append(r['reading'])
    if len(on) > 0:
        onyo.append(on)
    else:
        onyo.append(np.nan)
    if len(kun) > 0:
        kuyo.append(kun)
    else:
        kuyo.append(np.nan)
        
    mean=[]
    for m in temp_dat['meanings']:
        mean.append(m['meaning'])
    imi.append(mean)

print('Length Check: ',len(ids),len(WKlvl),len(char),len(imi),len(onyo),len(kuyo))
```

```python
df = pd.DataFrame([ids,WKlvl,char,imi,onyo,kuyo]).T
df.columns = ['ID','Level','Character','Meanings','Onyomi','Kunyomi']
df.index = df['ID']
df = df.drop(columns='ID')
df
```

```python
with open("WK_lvl10_test.txt",encoding="utf-8") as file:
    soup = file.read()

letters = "abcdefghijklmnopqrstuvwxyz"
caps = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
digits = "0123456789"
hira = """あいうえお
        かきくけこがぎぐげご
        さしすせそざじずぜぞ
        たちつてとだぢづでど
        なにぬねの
        はひふへほばびぶべぼぱぴぷぺぽ
        まみむめも
        やゃゆゅよょ
        らりるれろ
        わん"""

for c in soup:
    if c in letters:
        soup = soup.replace(c,"")
        continue
    if c in caps:
        soup = soup.replace(c,"")
        continue
    if c in digits:
        soup = soup.replace(c,"")
        continue
    if c in hira:
        soup = soup.replace(c,"")
        continue
        
soup = soup.replace("\n","")
soup = soup.replace("/","")
soup = soup.replace(" ","")
soup = soup.replace(",","")
soup = soup.replace("-","")
soup = soup.replace("©","")

soup
```
